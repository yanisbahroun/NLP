```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Here we are tokenize words. 
sentences = [ 
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think that my dog is amazing?'
]
tokenizer = Tokenizer(num_words = 100, oov_token = "<OOV>") #tokenize and only remember the 100 most frequent
# oov = out of vocabulary
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
print(word_index)
```

```python
# Now we need to tokenize "sentences", and process them to train NNs.
sequences = tokenizer.texts_to_sequences(sentences)
print(word_index)
print(sequences)
```

```python
# now what happens if I use the tokenizer on sentences with unseen words? when we add the oov token
test_data = [
    'I really love my dog',
    'my dog loves my manatee'
]

test_seq = tokenizer.texts_to_sequences(test_data)
print(test_seq)
```

```python
# What about if sequences of different lengths. We should use raggedtensors or padding. Here we will use padding
padded = pad_sequences(sequences)
print(word_index)
print(sequences)
print(padded) 
# it adds zeros at the beginning of the sentence until it will reach the longest sequence
# to put them at the front instead you can to padding = 'post' and set the maxlen=5
```

```python
padded = pad_sequences(sequences, padding = 'post',
                       truncating='post',maxlen=5)
print(word_index)
print(sequences)
print(padded) 
```
